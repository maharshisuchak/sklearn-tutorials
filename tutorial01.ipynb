## This is to learn basics of training a model and check the results of the same. We are using SVM basics for this

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import svm
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
digits = datasets.load_digits()
```

```python
(digits.data[:-1]).shape
```

```python
clf = svm.SVC(gamma=0.000001, C=100)

# Define the data and the values attached to it for the label
x,y = digits.data[:-10], digits.target[:-10] # :-10 means all data except last 10 from the data set.

# Train the model by putting data and label together
clf.fit(x,y)

# Run the model for a datum that was out of the training data
predicted=clf.predict(digits.data)[-2] #Running the model for last second datum from the data

# Prints the prediction of label for the image seen in the data at last second.
print('Prediction:',clf.predict(digits.data)[-2])

# Just prints the last second data and its image - in result you can see "9 9"
print(predicted,digits.target[-2])

plt.imshow(digits.images[-2], cmap=plt.cm.gray_r, interpolation="nearest")
plt.show()
```

```python
# Show the 10 data, image x and label y from all 100 training data

for i in range(10):
    image=np.reshape(digits.data[i],(8,8))
    print("Label",digits.target[i])
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    plt.show()
```
